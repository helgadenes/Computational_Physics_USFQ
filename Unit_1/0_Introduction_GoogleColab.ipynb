# 2. Python commands:


```python

# This # is used to add a comment in the python language

# Importing libraries

import numpy as np # np is an alias, aliases are used to avoid calling long names
import matplotlib.pyplot as plt # plt is an alias
```

### 2.1. 1D plot of a $\sin(x)$ function

Here we will plot:

$y =\sin(x)$

```python
# Let's make a plot

# To do inline plotting without an environment:
%matplotlib inline

# Generate a vector for x
x = np.arange(-np.pi, +np.pi, 0.01)

#print(x)

# Plotting (default)
plt.plot(x, np.sin(x))
```

*Plot* customisation:

```python
# Plotting with dotted line
plt.plot(x, np.sin(x), linestyle = ":", color = "red")
```

```python
# Plotting with markers
plt.plot(x, np.sin(x), linestyle = " ", marker = "+", color = "red")
```

### 2.2. 3D graphic of a surface

Let's make a 3D plot of the following surface:

$z = \sin(\sqrt{x^2+y^2})$

```python
# Let's define a function using def/return statements

def z_surface(x, y):
  z = np.sin(np.sqrt(x**2 + y**2))
  return z
```

```python
# We need to create a grid, for which we first generate 1D vectors
x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

#print(x)
#print(y)
```

```python
# We create a 2D Grid using the vectors above
X_2D, Y_2D = np.meshgrid(x, y)

# Checking the shapes of the arrays, we see they are 2D.
#print(X_2D.shape, Y_2D.shape)
```

```python
# We feed our function:
Z_2D = z_surface(X_2D, Y_2D)

# Checking the shape of the Z array
#print(Z_2D.shape)
```

```python
# Plotting with a figure environment
fig = plt.figure()

ax = plt.axes(projection = "3d")
ax.contour3D(X_2D, Y_2D, Z_2D, 200)

plt.show()
```
